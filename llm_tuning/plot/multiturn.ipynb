```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
# 
np.random.seed(0)

# 1
rounds = ['R1', 'R2', 'R3']
scores_method1_scene1 = [2.9, 2.87, 2.89]  # 1
scores_method2_scene1 = [2.9, 3.041, 3.25]  # 2
scores_method3_scene1 = [2.9, 3.01, 3.25]  # 3
# scores_method4_scene1 = [2.9, 3.12, 3.11]

# 2
scores_method1_scene2 = [3.65, 3.44, 3.34]  # 1
scores_method2_scene2 = [3.65, 3.49, 3.608]  # 2
scores_method3_scene2 = [3.65, 3.35, 3.278]  # 3
# scores_method4_scene2 = [3.65, 3.41, 3.24]  # 4

# 
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

marker_style = {'markerfacecolor': 'none', 'markersize': 10, 'markeredgewidth': 2}

# 1
axs[0].plot(rounds, scores_method1_scene1, label='Origin.', marker='o', **marker_style, color='#ffbe59', linewidth=2)
axs[0].plot(rounds, scores_method2_scene1, label='SFT', marker='s', **marker_style, color='#ff3e3e', linewidth=2)
axs[0].plot(rounds, scores_method3_scene1, label='PPO', marker='^', **marker_style, color='#39aeff', linewidth=2)
# axs[0].plot(rounds, scores_method4_scene1, label='RLHF', marker='D', **marker_style, color='#a25280', linewidth=2)
axs[0].set_ylim(2.6, 3.5)
axs[0].set_title('Job Seeking', fontsize=16, pad=10)
axs[0].set_xlabel('Rounds', fontsize=14)
axs[0].set_ylabel('Score', fontsize=14)
axs[0].legend(fontsize=12)
axs[0].set_yticks(np.round(np.linspace(2.6, 3.5, 5), 2))
axs[0].grid(True, linestyle='--', alpha=0.3, linewidth=1, color='gray')
axs[0].patch.set_facecolor('#f0f0f0')

# 2
axs[1].plot(rounds, scores_method1_scene2, label='Origin.', marker='o', **marker_style, color='#ffbe59', linewidth=2)
axs[1].plot(rounds, scores_method2_scene2, label='SFT', marker='s', **marker_style, color='#ff3e3e', linewidth=2)
axs[1].plot(rounds, scores_method3_scene2, label='PPO', marker='^', **marker_style, color='#39aeff', linewidth=2)
# axs[1].plot(rounds, scores_method4_scene2, label='RLHF', marker='D', **marker_style, color='#a25280', linewidth=2)
axs[1].set_ylim(3.1, 4.0)
axs[1].set_title('Recommendation', fontsize=16, pad=10)
axs[1].set_xlabel('Rounds', fontsize=14)
axs[1].set_ylabel('Score', fontsize=14)
axs[1].legend(fontsize=12)
axs[1].set_yticks(np.round(np.linspace(3.1, 4.0, 5), 2))
axs[1].grid(True, linestyle='--', alpha=0.3, linewidth=1, color='gray')
axs[1].patch.set_facecolor('#f0f0f0')

plt.tight_layout()

# pdf
plt.savefig('multi-rounds.pdf', format='pdf')   

plt.show()
```
